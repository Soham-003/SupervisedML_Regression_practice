```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = np.cos(x)
```

```python
plt.figure(figsize = (10,3))
plt.scatter(x,y,c='g')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("First Curve")
plt.show()
```

```python
sizes = np.random.rand(10)*100
colors = np.random.randn(10)
plt.figure(figsize=(8,8))
plt.scatter(x,y,s = sizes,c = colors)
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(x = df['sepal_length'],y = df['sepal_width'],hue = df.petal_length)
```

```python
df['petal_length'].unique()
```

```python
sns.histplot(df.petal_length,kde = True,color='g')
```

```python
sns.relplot(data = df,x = "sepal_length",y = "sepal_width",style = 'petal_length',hue = 'petal_length')
```
